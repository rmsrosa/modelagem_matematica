# O Pêndulo

Nesse capítulo, vamos considerar o movimento de um pêndulo, que consiste em uma objeto preso a uma extremidade de uma haste, com o outro extremo desta haste preso a um determinado ponto fixo. É considerado o caso de um objeto com massa bem maior que a da haste. Permite-se, então, ao objeto, balançar, sob a ação da gravidade. A haste é escolhida com baixa elasticidade, de forma que o movimento é feito de tal forma que o objeto se mantenha praticamente em uma esfera de centro no ponto fixo e de raio igual ao comprimento da haste.

Na situação de interesse, tratamos do caso em que o movimento inicial é tal que o movimento do pêndulo se dê de quase de forma planar, ou seja, quase restrito à uma circunferência.
