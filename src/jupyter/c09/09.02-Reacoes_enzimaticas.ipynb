```julia
using Random
using Statistics
using LinearAlgebra
using LsqFit
using CSV
using Plots
theme(:ggplot2)
```

```julia
S = [csv_data[j][1] for j in 1:length(csv_data)]
v = [csv_data[j][2] for j in 1:length(csv_data)]
scatter(S,v, xlabel="S", ylabel="v", label=false,
    title = "Enzima vs velocidade", titlefont=10)
```
